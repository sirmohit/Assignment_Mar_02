Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans:
    
    What is Matplotlib?
    Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in Python.
    It provides a wide variety of plotting functionalities and is highly customizable, allowing users to create publication
    -quality figures. Matplotlib is built on the concept of providing low-level, high-level, and object-oriented APIs for
    creating visualizations, making it flexible for various plotting needs.

    Why is it used?
    Data Visualization: Matplotlib is primarily used for visualizing data in a graphical format, enabling users to explore
        and understand complex datasets more effectively.
    Publication-Quality Graphics: It provides tools to create high-quality figures suitable for publications, presentations,
        and reports.
    Customizability: Matplotlib offers extensive customization options for controlling every aspect of the plot, including 
        colors, styles, annotations, and layouts.
    Compatibility: It integrates well with other libraries and frameworks in the Python ecosystem, such as NumPy, Pandas,
        and Seaborn, making it versatile for data analysis and visualization tasks.
    Flexibility: Matplotlib supports various plot types, styles, and formats, allowing users to create a wide range of
        visualizations tailored to specific requirements.
        
    Five plots that can be plotted using the Pyplot module of Matplotlib:
    Line Plot: Used to visualize the relationship between two variables by plotting data points connected by straight lines.
    Scatter Plot: Represents individual data points as markers on a two-dimensional plane, suitable for visualizing 
        relationships between two numerical variables.
    Histogram: Displays the distribution of a single numerical variable by dividing the data into bins and plotting the 
        frequency or count of observations in each bin.
    Bar Chart: Represents categorical data using rectangular bars, where the length or height of the bars corresponds to 
        the frequency, count, or value of each category.
    Pie Chart: Displays categorical data as a circular graph, where each slice represents a category, and the size of the 
        slice corresponds to the proportion or percentage of the category relative to the whole.


Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
'''
A scatter plot is a type of data visualization that represents individual data points as markers on a two-dimensional
plane, with one variable plotted along the x-axis and another variable plotted along the y-axis. Scatter plots are
particularly useful for visualizing relationships or correlations between two numerical variables, enabling you to identify
patterns, trends, clusters, or outliers in the data.'''

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.scatter(x, y, color='blue', alpha=0.7)  # Create scatter plot with transparency

# Add title, xlabel, and ylabel
plt.title('Scatter Plot of X vs. Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.grid(True)  # Add gridlines
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define data arrays for each line plot
x = np.array([0, 1, 2, 3, 4, 5])

# Define y-values for each line plot
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and arrange four subplots using subplot()
plt.figure(figsize=(12, 8))  # Set figure size

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, y1, marker='o', color='b')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot
plt.plot(x, y2, marker='s', color='r')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot
plt.plot(x, y3, marker='*', color='g')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot
plt.plot(x, y4, marker='d', color='m')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
'''A bar plot is a type of data visualization that represents categorical data using rectangular bars, where the length or
height of each bar corresponds to the frequency, count, or value of the respective category. Bar plots are particularly useful
for comparing and visualizing the differences between discrete categories, making it easier to identify trends, rankings, or
variations in the data.
Comparison: Bar plots provide a visual representation of categorical data, allowing users to compare the values or frequencies
of different categories easily.
Ranking: They are useful for ranking categories based on their values or frequencies, highlighting the highest or lowest
values among the categories.
Trends and Patterns: Bar plots help in identifying trends, patterns, or variations in categorical data, enabling users to
make informed decisions or interpretations.
Presentation: They are commonly used in presentations, reports, and publications to convey information effectively and engage
the audience.'''

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define data arrays for company and profit
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.bar(company, profit, color='skyblue')
plt.title('Vertical Bar Plot of Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.grid(True)  # Add gridlines

# Display the vertical bar plot
plt.tight_layout()
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.barh(company, profit, color='salmon')
plt.title('Horizontal Bar Plot of Company Profits')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.grid(True)  # Add gridlines

# Display the horizontal bar plot
plt.tight_layout()
plt.show()

```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
'''A box plot, also known as a whisker plot, is a type of data visualization that provides a graphical representation of
the distribution of a dataset. It displays key statistical measures, including the median, quartiles, and potential outliers,
using a box-and-whisker diagram. The box encompasses the interquartile range (IQR), which represents the middle 50% of 
the data, while the whiskers extend to show the range of the data excluding potential outliers.

Why is it used:

Summary of Data Distribution: A box plot summarizes the distribution of a dataset by highlighting key statistical measures,
such as the median, quartiles, and potential outliers, providing insights into the central tendency and spread of the data.

Comparison: It is useful for comparing the distributions of multiple datasets or groups, enabling users to identify 
variations, similarities, or differences between the groups.
Identification of Outliers: Box plots help in identifying potential outliers or extreme values in the dataset, facilitating 
data cleaning and analysis.
Visualizing Skewness and Spread: They are effective for visualizing skewness, spread, and variability in the data, making
it easier to interpret and understand the underlying patterns or trends.
'''
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate data arrays for box plots
box1 = np.random.normal(100, 10, 200)  # Mean = 100, Standard Deviation = 10
box2 = np.random.normal(90, 20, 200)   # Mean = 90, Standard Deviation = 20

# Combine data into a single list for plotting
data_to_plot = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))  # Set figure size
plt.boxplot(data_to_plot, labels=['Box 1', 'Box 2'], patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box Plot of Box1 and Box2')
plt.xlabel('Box')
plt.ylabel('Value')
plt.grid(True)  # Add gridlines

# Display the box plot
plt.tight_layout()
plt.show()
```
